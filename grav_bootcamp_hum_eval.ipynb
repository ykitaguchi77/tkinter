```python
import tkinter as tk
import csv
import time

def submit():
    elapsed_time = time.time() - start_time
    with open("output.csv", "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["elapsed time", elapsed_time])
        writer.writerows(answers)
    label.config(text="入力が完了しました")

def yes():
    global index
    answers[index] = [index, "yes"]
    index += 1
    label.config(text=f"画像{index}について評価してください")
    if index == 20:
        submit_button.config(state="normal")
        yes_button.config(state="disabled")
        no_button.config(state="disabled")
        label.config(text=f"submitを押して終了してください。お疲れ様でした！")

def no():
    global index
    answers[index] = [index, "no"]
    index += 1
    label.config(text=f"画像{index}について評価してください")
    if index == 20:
        submit_button.config(state="normal")
        yes_button.config(state="disabled")
        no_button.config(state="disabled") 
        label.config(text=f"submitを押して終了してください。お疲れ様でした！")

def back():
    global index
    index -= 1
    answers[index] = [index, ""]
    label.config(text=f"画像{index}について評価してください")

def start_evaluation():
    global start_time
    start_time = time.time()
    name = name_entry.get()
    label.config(text=f"{name}さん、画像{index}について評価してください")
    name_entry.config(state="disabled")
    start_button.config(state="disabled")
    yes_button.config(state="normal")
    no_button.config(state="normal")

root = tk.Tk()
root.title("評価アプリ")

index = 0
answers = [[i, ""] for i in range(20)]

label = tk.Label(root, text="名前を入力してください")
label.pack()

name_entry = tk.Entry(root)
name_entry.pack()

start_button = tk.Button(root, text="評価を開始する", command=start_evaluation)
start_button.pack(side="bottom")

label = tk.Label(root, text=f"画像{index}について評価してください")
label.pack()

yes_button = tk.Button(root, text="はい", command=yes)
yes_button.pack(side="left")

no_button = tk.Button(root, text="いいえ", command=no)
no_button.pack(side="left")

back_button = tk.Button(root, text="前の項目に戻る", state="normal", command=back)
back_button.pack(side="left")

submit_button = tk.Button(root, text="submit", state="disabled", command=submit)
submit_button.pack(side="bottom")

quit_button = tk.Button(root, text="強制終了", command=root.destroy)
quit_button.pack(side="bottom")

root.protocol("WM_DELETE_WINDOW", root.destroy)
root.mainloop()
```
